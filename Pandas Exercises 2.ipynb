# Given the following Code

```python
import pandas as pd
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', 
                    sep='|', index_col='user_id')
```

# Print out a sample of the dataframe for viewing

```python

users.sample(10)


```

# Find the mean age per occupation

```python
users.groupby('occupation').mean()



```

# Find the mean age per gender

```python
users.groupby('gender').mean()



```

# Print out the total for each occupation

```python


users[].groupby('occupation').size()

```

# Find the total male per occupation

```python

users[users['gender'] == 'M'].groupby('occupation').size().sort_values(ascending=False)

```

# Display the male ratio per occupation (in descending order)

```python


male_ratio_by_occup = (users[users['gender'] == 'M'].groupby('occupation').size())/users.groupby('occupation').size()*100
print(male_ratio_by_occup.sort_values(ascending=False).round(2))
```

# Display the female ratio per occupation (in descending order)

```python



male_ratio_by_occup = (users[users['gender'] == 'F'].groupby('occupation').size())/users.groupby('occupation').size()*100
print(male_ratio_by_occup.sort_values(ascending=False).round(2))

```

# For each occupation, find the min, mean, and max age

```python

users.groupby('occupation').describe()

```

# For each occupation find the total of women and men

```python
users.groupby(['occupation', 'gender']).agg({'zip_code': "count"})
```

# For each occupation find the percentage of women and men

```python

occup_gender_cnt = users.groupby(['occupation', 'gender']).agg({'zip_code': "count"})

occup_gender_cnt.groupby(level=0).apply(lambda x: 100 * x / float(x.sum()))
```

# For each occupation find the total of women and men and the min, mean, and max age for each gender

```python

users.groupby(['occupation', 'gender']).describe()
```
